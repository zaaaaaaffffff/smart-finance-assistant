### Foundation Data Processing Functions

````python
"""
transaction_loader.py
---------------------
Load and clean financial transaction CSV data for the Smart Finance Assistant.
"""

from __future__ import annotations

import io
import re

import pandas as pd

# ---------------------------------------------------------------------------
# Constants
# ---------------------------------------------------------------------------

REQUIRED_COLUMNS: frozenset[str] = frozenset({"Date", "Amount", "Category", "Description"})

# Optional: normalise free-text categories to canonical names.
# Set to None (or remove) to skip normalisation entirely.
CATEGORY_ALIASES: dict[str, str] = {
    "food": "Food",
    "groceries": "Food",
    "dining": "Food",
    "transport": "Transport",
    "transportation": "Transport",
    "travel": "Transport",
    "shopping": "Shopping",
    "clothes": "Shopping",
    "refund": "Refund",
    "income": "Income",
    "salary": "Income",
}

# Pre-compiled pattern: strips $, commas, and converts (123.45) → -123.45
_AMOUNT_PATTERN = re.compile(r"[$,]")
_PARENS_PATTERN = re.compile(r"^\((.+)\)$")


# ---------------------------------------------------------------------------
# Private helpers
# ---------------------------------------------------------------------------

def _clean_amount_series(series: pd.Series) -> pd.Series:
    """Convert a raw Amount column (strings) to float64.

    Handles:
      - Dollar signs and commas:   "$1,234.56" → 1234.56
      - Parentheses negatives:     "(45.00)"   → -45.00
      - Explicit minus:            "-$40"      → -40.0
    """
    s = series.astype(str).str.strip()
    # Replace parentheses notation with explicit negative sign
    s = s.str.replace(_PARENS_PATTERN, r"-\1", regex=True)
    # Strip currency symbols and thousands separators
    s = s.str.replace(_AMOUNT_PATTERN, "", regex=True)
    return pd.to_numeric(s, errors="coerce")


def _normalise_categories(series: pd.Series) -> pd.Series:
    """Map free-text categories to canonical names using CATEGORY_ALIASES."""
    return series.str.strip().apply(
        lambda v: CATEGORY_ALIASES.get(v.lower(), v) if isinstance(v, str) else v
    )


# ---------------------------------------------------------------------------
# Public API
# ---------------------------------------------------------------------------

def load_and_clean_transaction_data(
    file_path: str | io.IOBase,
    date_format: str | None = None,
    normalise_categories: bool = True,
) -> pd.DataFrame:
    """Load and clean transaction data from a CSV file.

    Expected columns: Date, Amount, Category, Description.

    Args:
        file_path:            Path to a CSV file, or a file-like object (e.g. io.StringIO).
        date_format:          Optional strftime format string for the Date column
                              (e.g. ``"%Y-%m-%d"``). Pass ``None`` to let pandas
                              infer the format (slower but more flexible).
        normalise_categories: When True, map category aliases to canonical names
                              via ``CATEGORY_ALIASES``.

    Returns:
        A cleaned ``pd.DataFrame`` with columns:
        ``Date`` (datetime64), ``Amount`` (float64), ``Category`` (str),
        ``Description`` (str).

    Raises:
        Exception: With a user-friendly message for all anticipated error cases.
    """
    try:
        # ------------------------------------------------------------------
        # 1. Load
        # ------------------------------------------------------------------
        df = pd.read_csv(file_path)

        if df.empty:
            raise ValueError("The uploaded file contains no data rows.")

        # Strip accidental whitespace from column headers
        df.columns = df.columns.str.strip()

        # ------------------------------------------------------------------
        # 2. Validate schema
        # ------------------------------------------------------------------
        missing_cols = REQUIRED_COLUMNS - set(df.columns)
        if missing_cols:
            raise KeyError(
                f"Missing required columns: {', '.join(sorted(missing_cols))}. "
                "Please ensure Date, Amount, Category, and Description are present."
            )

        # ------------------------------------------------------------------
        # 3. Drop rows where the two critical fields are entirely absent
        # ------------------------------------------------------------------
        df = df.dropna(subset=["Date", "Amount"])

        # ------------------------------------------------------------------
        # 4. Fill optional fields
        # ------------------------------------------------------------------
        df["Category"] = df["Category"].fillna("Unknown").astype(str)
        df["Description"] = df["Description"].fillna("Unknown").astype(str).str.strip()

        # ------------------------------------------------------------------
        # 5. Parse Amount
        # ------------------------------------------------------------------
        df["Amount"] = _clean_amount_series(df["Amount"])

        # ------------------------------------------------------------------
        # 6. Parse Date
        # ------------------------------------------------------------------
        df["Date"] = pd.to_datetime(df["Date"], format=date_format, errors="coerce")

        # ------------------------------------------------------------------
        # 7. Drop rows that failed numeric / date coercion
        # ------------------------------------------------------------------
        df = df.dropna(subset=["Date", "Amount"])

        if df.empty:
            raise ValueError(
                "No valid records remained after cleaning. "
                "Check that Amount values are numeric and Date values are parseable."
            )

        # ------------------------------------------------------------------
        # 8. Optional category normalisation
        # ------------------------------------------------------------------
        if normalise_categories:
            df["Category"] = _normalise_categories(df["Category"])

        return df.reset_index(drop=True)

    except FileNotFoundError:
        raise Exception("Error: The transaction file could not be found at the provided path.")
    except pd.errors.EmptyDataError:
        raise Exception("Error: The uploaded CSV file is blank or corrupted.")
    except (KeyError, ValueError) as business_err:
        raise Exception(f"Validation Error: {business_err}")
    except Exception as general_err:
        raise Exception(f"Unexpected error during file processing: {general_err}")


# ---------------------------------------------------------------------------
# Tests
# ---------------------------------------------------------------------------

def _run_tests() -> None:
    # ------------------------------------------------------------------
    # Test 1: Happy path — all edge cases in Amount / Date / Category
    # ------------------------------------------------------------------
    valid_csv = """Date,Amount,Category,Description
2026-05-04,$45,Food,Lunch
2026-05-05,$30,Transport,Bus Pass
2026-05-06,$75,dining,Dinner out
2026-05-07,$100,Shopping,Clothes
2026-05-08,-$40,Refund,Returned item
2026-05-09,($15.50),Food,Late Night Fast Food
, $20, Food, Missing Date Row (Should drop)
2026-05-10,,Shopping,Missing Amount Row (Should drop)
2026-05-11,$12,,Missing Category Row (Should become Unknown)
"""
    print("--- Test 1: Valid data with mixed Amount formats ---")
    df = load_and_clean_transaction_data(io.StringIO(valid_csv), date_format="%Y-%m-%d")
    print(df.to_string())
    print(f"\nDate dtype : {df['Date'].dtype}")
    print(f"Amount dtype: {df['Amount'].dtype}")

    assert df["Date"].dtype == "datetime64[ns]", "Date column must be datetime"
    assert df["Amount"].dtype == "float64", "Amount column must be float64"
    assert len(df) == 7, f"Expected 7 rows after dropping invalids, got {len(df)}"
    assert df.loc[df["Description"] == "Late Night Fast Food", "Amount"].iloc[0] == -15.5
    assert df.loc[df["Description"] == "Missing Category Row (Should become Unknown)", "Category"].iloc[0] == "Unknown"
    # Category alias: "dining" → "Food"
    assert df.loc[df["Description"] == "Dinner out", "Category"].iloc[0] == "Food"
    print("\n✓ Test 1 passed\n")

    # ------------------------------------------------------------------
    # Test 2: Schema mismatch → should raise
    # ------------------------------------------------------------------
    invalid_csv = """TransactionDate,CashValue,Tag
2026-05-04,$45,Food
2026-05-05,$30,Transport
"""
    print("--- Test 2: Missing required columns ---")
    try:
        load_and_clean_transaction_data(io.StringIO(invalid_csv))
        assert False, "Should have raised"
    except Exception as e:
        print(f"Caught expected error: {e}")
        assert "Missing required columns" in str(e)
    print("✓ Test 2 passed\n")

    # ------------------------------------------------------------------
    # Test 3: Empty CSV → should raise
    # ------------------------------------------------------------------
    empty_csv = "Date,Amount,Category,Description\n"
    print("--- Test 3: Empty CSV (headers only) ---")
    try:
        load_and_clean_transaction_data(io.StringIO(empty_csv))
        assert False, "Should have raised"
    except Exception as e:
        print(f"Caught expected error: {e}")
        assert "no data rows" in str(e).lower() or "no valid records" in str(e).lower()
    print("✓ Test 3 passed\n")

    print("All tests passed.")


if __name__ == "__main__":
    _run_tests()
````

````python
"""
spending_analysis.py
--------------------
Analyse spending patterns from a cleaned transaction DataFrame and produce
structured results suitable for both programmatic consumption and display.
"""

from __future__ import annotations

from dataclasses import dataclass, field
from typing import NamedTuple

import pandas as pd

# ---------------------------------------------------------------------------
# Configuration — edit these thresholds without touching business logic
# ---------------------------------------------------------------------------

FIXED_CATEGORIES: frozenset[str] = frozenset({"Utilities", "Rent", "Insurance", "Bills"})
DISCRETIONARY_CATEGORIES: frozenset[str] = frozenset(
    {"Food", "Dining Out", "Shopping", "Entertainment", "Retail"}
)

FIXED_COST_WARNING_PCT: float = 50.0       # warn if fixed costs exceed this % of gross spend
DISCRETIONARY_SPIKE_PCT: float = 25.0      # warn if a single discretionary category exceeds this %
TOP_N_CATEGORIES: int = 3


# ---------------------------------------------------------------------------
# Result types — callers get structured data, not just a markdown string
# ---------------------------------------------------------------------------

class SpendSummary(NamedTuple):
    gross_spend: float
    total_refunds: float          # always ≤ 0
    net_spend: float


@dataclass
class SpendingAnalysis:
    """All outputs from :func:`analyze_spending_patterns`."""
    summary: SpendSummary
    category_breakdown: pd.DataFrame   # columns: Total Spend, Transaction Count, Percentage
    top_categories: list[str]
    insights: list[str]
    summary_markdown: str = field(repr=False)


# ---------------------------------------------------------------------------
# Private helpers
# ---------------------------------------------------------------------------

def _build_category_breakdown(spending_df: pd.DataFrame, total_gross: float) -> pd.DataFrame:
    """Aggregate by category and attach percentage column."""
    group = (
        spending_df.groupby("Category")["Amount"]
        .agg(total_spend="sum", transaction_count="count")
        .rename(columns={"total_spend": "Total Spend", "transaction_count": "Transaction Count"})
    )
    group["Percentage"] = (group["Total Spend"] / total_gross) * 100
    return group.sort_values("Total Spend", ascending=False)


def _generate_insights(
    breakdown: pd.DataFrame,
    summary: SpendSummary,
) -> list[str]:
    """Return a list of plain-text insight strings (no markdown inside)."""
    insights: list[str] = []
    gross = summary.gross_spend

    # --- Fixed-cost health check ---
    fixed_spend = breakdown.loc[breakdown.index.isin(FIXED_CATEGORIES), "Total Spend"].sum()
    fixed_pct = (fixed_spend / gross) * 100
    if fixed_pct > FIXED_COST_WARNING_PCT:
        insights.append(
            f"Fixed costs ({fixed_pct:.1f}% of gross spend) exceed the recommended {FIXED_COST_WARNING_PCT:.0f}% "
            "ceiling. Review subscriptions and utility plans for savings opportunities."
        )
    else:
        insights.append(
            f"Fixed costs are healthy at {fixed_pct:.1f}% of gross spend, "
            "leaving adequate room for savings and discretionary flexibility."
        )

    # --- Discretionary spike detection ---
    for cat in DISCRETIONARY_CATEGORIES:
        if cat in breakdown.index:
            cat_pct = breakdown.loc[cat, "Percentage"]
            if cat_pct > DISCRETIONARY_SPIKE_PCT:
                insights.append(
                    f"{cat} accounts for {cat_pct:.1f}% of gross spend — "
                    f"above the {DISCRETIONARY_SPIKE_PCT:.0f}% threshold. "
                    "Consider a temporary budget cap on this category."
                )

    # --- Refund efficiency ---
    if summary.total_refunds < 0:
        refund_pct = (abs(summary.total_refunds) / gross) * 100
        insights.append(
            f"Refunds recovered ${abs(summary.total_refunds):,.2f}, "
            f"offsetting {refund_pct:.1f}% of gross spend."
        )

    return insights


def _render_markdown(
    summary: SpendSummary,
    breakdown: pd.DataFrame,
    top_categories: list[str],
    insights: list[str],
) -> str:
    """Render analysis results as a markdown report string."""
    lines: list[str] = [
        "## Smart Finance Assistant — Spending Report",
        "",
        "### Core Metrics",
        f"| Metric | Value |",
        f"| :--- | ---: |",
        f"| Gross spend | ${summary.gross_spend:,.2f} |",
        f"| Total refunds | ${abs(summary.total_refunds):,.2f} |",
        f"| Net spend | ${summary.net_spend:,.2f} |",
        "",
        "### Expenditure by Category",
        "| Category | Total | Transactions | % of Gross |",
        "| :--- | ---: | :---: | ---: |",
    ]

    for cat, row in breakdown.iterrows():
        lines.append(
            f"| {cat} | ${row['Total Spend']:,.2f} | {int(row['Transaction Count'])} | {row['Percentage']:.1f}% |"
        )

    lines += [
        "",
        f"**Top {TOP_N_CATEGORIES} categories:** {', '.join(top_categories)}",
        "",
        "### Insights",
    ]
    for insight in insights:
        lines.append(f"- {insight}")

    return "\n".join(lines)


# ---------------------------------------------------------------------------
# Public API
# ---------------------------------------------------------------------------

def analyze_spending_patterns(df: pd.DataFrame) -> SpendingAnalysis:
    """Analyse spending patterns and return a structured :class:`SpendingAnalysis`.

    Args:
        df: A cleaned transaction DataFrame with at minimum the columns
            ``Amount`` (float64) and ``Category`` (str). Positive amounts
            represent outflows; negative amounts represent refunds/credits.

    Returns:
        A :class:`SpendingAnalysis` dataclass containing the category
        breakdown, top categories, insights list, and a pre-rendered
        markdown report.

    Raises:
        ValueError: If ``df`` is missing required columns.
    """
    required = {"Amount", "Category"}
    missing = required - set(df.columns)
    if missing:
        raise ValueError(f"DataFrame is missing required columns: {', '.join(sorted(missing))}")

    spending_df = df[df["Amount"] > 0].copy()
    refund_df = df[df["Amount"] < 0].copy()

    gross = spending_df["Amount"].sum()
    refunds = refund_df["Amount"].sum()          # ≤ 0
    net = df["Amount"].sum()

    summary = SpendSummary(gross_spend=gross, total_refunds=refunds, net_spend=net)

    if gross == 0:
        empty_breakdown = pd.DataFrame(
            columns=["Total Spend", "Transaction Count", "Percentage"]
        )
        return SpendingAnalysis(
            summary=summary,
            category_breakdown=empty_breakdown,
            top_categories=[],
            insights=["No positive spending records found for this period."],
            summary_markdown="## Smart Finance Assistant\n\nNo positive spending records found.",
        )

    breakdown = _build_category_breakdown(spending_df, gross)
    top_categories = breakdown.index.tolist()[:TOP_N_CATEGORIES]
    insights = _generate_insights(breakdown, summary)
    markdown = _render_markdown(summary, breakdown, top_categories, insights)

    return SpendingAnalysis(
        summary=summary,
        category_breakdown=breakdown,
        top_categories=top_categories,
        insights=insights,
        summary_markdown=markdown,
    )


# ---------------------------------------------------------------------------
# Tests
# ---------------------------------------------------------------------------

def _run_tests() -> None:
    base_data = pd.DataFrame([
        {"Amount": 120.00, "Category": "Food",          "Description": "Groceries"},
        {"Amount":  60.00, "Category": "Transport",     "Description": "Taxi"},
        {"Amount": 200.00, "Category": "Entertainment", "Description": "Concert"},
        {"Amount": 400.00, "Category": "Rent",          "Description": "Monthly rent"},
        {"Amount": -50.00, "Category": "Refund",        "Description": "Return"},
    ])

    print("--- Test 1: Standard analysis ---")
    result = analyze_spending_patterns(base_data)

    assert isinstance(result, SpendingAnalysis)
    assert result.summary.gross_spend == 780.0
    assert result.summary.total_refunds == -50.0
    assert result.summary.net_spend == 730.0
    assert result.top_categories[0] == "Rent"
    assert len(result.insights) >= 1
    assert "Rent" in result.category_breakdown.index
    print("Insights generated:")
    for insight in result.insights:
        print(f"  - {insight}")
    print("✓ Test 1 passed\n")

    print("--- Test 2: No spending (refunds only) ---")
    refund_only = pd.DataFrame([{"Amount": -30.0, "Category": "Refund", "Description": "Return"}])
    result2 = analyze_spending_patterns(refund_only)
    assert result2.top_categories == []
    assert result2.category_breakdown.empty
    print("✓ Test 2 passed\n")

    print("--- Test 3: Missing required column raises ValueError ---")
    try:
        analyze_spending_patterns(pd.DataFrame([{"Amount": 100}]))
        assert False, "Should have raised"
    except ValueError as e:
        assert "Category" in str(e)
    print("✓ Test 3 passed\n")

    print("--- Test 4: Discretionary spike triggers insight ---")
    spike_data = pd.DataFrame([
        {"Amount": 500.0, "Category": "Food",      "Description": "Bulk order"},
        {"Amount":  50.0, "Category": "Transport", "Description": "Bus"},
    ])
    result4 = analyze_spending_patterns(spike_data)
    assert any("Food" in i for i in result4.insights), "Expected Food spike insight"
    print("✓ Test 4 passed\n")

    print("All tests passed.")


if __name__ == "__main__":
    _run_tests()
````

````python

"""
finance_chatbot.py
------------------
Budget Buddy: AI-powered personal finance chat assistant for Google Colab.
"""

from hands_on_ai.chat import get_response
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output
import datetime
from dataclasses import dataclass, field


# ── System prompt builder ─────────────────────────────────────

def build_system_prompt(analysis_data: dict | None = None) -> str:
    """
    Build a system prompt, optionally injecting real spending context
    for data-aware, personalised financial guidance.
    """
    base = """
You are Penny, a savvy and supportive personal finance coach.

## Personality
- Friendly, calm, and judgment-free — finances can be overwhelming, keep it simple
- Action-oriented: every response must include something the user can do today
- Straight-talking: if something looks risky, say it kindly but clearly
- Brief: use bullet points or short paragraphs, never long walls of text

## Expertise
- Personal budgeting and cash-flow planning
- Spending category analysis and optimisation
- Savings habits (emergency funds, sinking funds, automation tips)
- Debt paydown strategies (avalanche and snowball methods)
- Recurring charge and subscription audits
- Understanding how refunds affect your real net spend

## Response rules
- Always finish with one clear, specific action the user can take right now
- Never recommend specific financial products, investments, or securities
- If asked something unrelated to personal finance, redirect politely
- Define any financial jargon you use
- Keep replies under 200 words unless the user requests more detail
""".strip()

    if not analysis_data:
        return base

    currency        = analysis_data.get("currency", "AUD")
    total_spending  = analysis_data.get("total_spending", 0)
    net_spending    = analysis_data.get("net_spending", 0)
    avg_transaction = analysis_data.get("avg_transaction", 0)
    tx_count        = analysis_data.get("transaction_count", 0)
    top_cats        = analysis_data.get("top_categories", {})
    pct             = analysis_data.get("pct_by_category", {})
    monthly         = analysis_data.get("monthly_totals", {})
    largest         = analysis_data.get("largest_transaction", {})

    cats_text = "\n".join(
        f"  - {cat}: {currency} ${val:,.2f} ({pct.get(cat, 0):.1f}%)"
        for cat, val in top_cats.items()
    ) or "  - No category data available"

    monthly_text = "\n".join(
        f"  - {m}: {currency} ${v:,.2f}" for m, v in sorted(monthly.items())
    ) or "  - No monthly data available"

    largest_text = (
        f"  {largest.get('description', 'N/A')} — "
        f"{currency} ${largest.get('amount', 0):,.2f} "
        f"({largest.get('category', '')}, {largest.get('date', '')})"
        if largest else "  None recorded"
    )

    context = f"""

## User's Spending Data (from uploaded CSV)
Reference this naturally in conversation ("I can see your top category is…").
Focus on what the numbers mean for their financial wellbeing, not just the figures.

- Gross spend     : {currency} ${total_spending:,.2f}
- Net spend       : {currency} ${net_spending:,.2f}
- Avg transaction : {currency} ${avg_transaction:,.2f}
- Transactions    : {tx_count}

Top categories:
{cats_text}

Monthly breakdown:
{monthly_text}

Largest transaction:
{largest_text}
"""
    return base + context


# ── Chat session ──────────────────────────────────────────────

@dataclass
class FinanceChatSession:
    """
    Manages conversation history and calls hands_on_ai get_response()
    with a persistent system prompt.
    """

    STARTERS = [
        "Where am I overspending?",
        "How do I start an emergency fund?",
        "Which category should I cut first?",
        "How do refunds affect my budget?",
        "How can I reduce my biggest expense?",
        "What's a simple way to save more?",
    ]

    analysis_data: dict | None = None
    history: list[dict] = field(default_factory=list)
    _system_prompt: str = field(init=False)

    def __post_init__(self):
        self._system_prompt = build_system_prompt(self.analysis_data)

    def _ts(self) -> str:
        return datetime.datetime.now().strftime("%H:%M")

    def _bubble(self, role: str, text: str, time: str) -> str:
        is_user   = role == "user"
        align     = "flex-end"  if is_user else "flex-start"
        bg        = "#2563eb"   if is_user else "#f0fdf4"
        color     = "#ffffff"   if is_user else "#14532d"
        label     = "You"       if is_user else "Penny 💚"
        radius    = "16px 16px 4px 16px" if is_user else "16px 16px 16px 4px"
        side      = "right"     if is_user else "left"
        safe_text = text.replace("\n", "<br>")
        return f"""
        <div style="display:flex;justify-content:{align};margin-bottom:10px">
          <div style="max-width:75%">
            <div style="font-size:11px;color:#9ca3af;margin-bottom:2px;text-align:{side}">
              {label} · {time}
            </div>
            <div style="background:{bg};color:{color};padding:10px 14px;
                        border-radius:{radius};font-family:sans-serif;
                        font-size:14px;line-height:1.5">
              {safe_text}
            </div>
          </div>
        </div>"""

    def _render(self, output_area) -> None:
        with output_area:
            clear_output(wait=True)
            if not self.history:
                display(HTML("""
                <div style="font-family:sans-serif;color:#9ca3af;text-align:center;
                            padding:40px;font-size:14px">
                  Ask Penny anything about your finances 💬
                </div>"""))
                return
            html = "".join(
                self._bubble(m["role"], m["text"], m["time"]) for m in self.history
            )
            display(HTML(f'<div style="padding:8px">{html}</div>'))

    def send(self, message: str, output_area=None) -> str:
        message = message.strip()
        if not message:
            return ""

        self.history.append({"role": "user", "text": message, "time": self._ts()})
        if output_area:
            self._render(output_area)

        try:
            response = get_response(message, system=self._system_prompt)
        except TypeError:
            try:
                response = get_response(message, self._system_prompt)
            except Exception as e:
                response = f"⚠️ Could not reach the AI model. ({e})"
        except Exception as e:
            response = f"⚠️ Something went wrong. ({e})"

        self.history.append({"role": "assistant", "text": response, "time": self._ts()})
        if output_area:
            self._render(output_area)
        return response

    def clear(self) -> None:
        """Reset conversation history while keeping the system prompt."""
        self.history.clear()


# ── Colab UI ──────────────────────────────────────────────────

def launch_finance_chat(analysis_data: dict | None = None) -> FinanceChatSession:
    """
    Launch the Budget Buddy chat widget in Google Colab.

    Parameters
    ----------
    analysis_data : dict from your spending analyser — optional but recommended
                    for personalised responses.

    Returns
    -------
    FinanceChatSession — can also call session.send() programmatically.
    """
    session = FinanceChatSession(analysis_data=analysis_data)

    has_data    = bool(analysis_data)
    badge       = ("✅ Spending data loaded — Penny knows your numbers!"
                   if has_data else
                   "ℹ️ No spending data loaded — Penny will give general advice")
    badge_color = "#10b981" if has_data else "#f59e0b"

    # Header
    display(HTML(f"""
    <div style="font-family:sans-serif;border:1px solid #d1fae5;border-radius:12px;
                overflow:hidden;margin-bottom:8px">
      <div style="background:#16a34a;padding:14px 18px;color:#fff;
                  display:flex;align-items:center;gap:10px">
        <span style="font-size:22px">💰</span>
        <div>
          <div style="font-weight:700;font-size:16px">Penny — Your Finance Coach</div>
          <div style="font-size:12px;opacity:.8">Smart Finance Assistant</div>
        </div>
      </div>
      <div style="background:#f0fdf4;padding:8px 16px;font-size:12px;color:{badge_color}">
        {badge}
      </div>
    </div>
    """))

    # Conversation area
    output_area = widgets.Output(
        layout=widgets.Layout(
            height="340px", overflow_y="auto",
            border="1px solid #d1fae5", border_radius="8px", padding="4px",
        )
    )
    session._render(output_area)
    display(output_area)

    # Starter chips
    chips = [
        widgets.Button(
            description=q,
            layout=widgets.Layout(width="auto", height="30px"),
            style=widgets.ButtonStyle(button_color="#dcfce7", font_color="#16a34a"),
        )
        for q in FinanceChatSession.STARTERS
    ]
    display(widgets.HBox(
        chips,
        layout=widgets.Layout(flex_wrap="wrap", gap="6px", margin="8px 0")
    ))

    # Input row
    text_input = widgets.Text(
        placeholder="Ask Penny about your finances…",
        layout=widgets.Layout(flex="1", height="36px"),
    )
    send_btn  = widgets.Button(
        description="Send", button_style="success",
        layout=widgets.Layout(width="80px", height="36px"),
    )
    clear_btn = widgets.Button(
        description="Clear",
        layout=widgets.Layout(width="70px", height="36px"),
        style=widgets.ButtonStyle(button_color="#f3f4f6"),
    )
    display(widgets.HBox(
        [text_input, send_btn, clear_btn],
        layout=widgets.Layout(gap="6px", margin="4px 0"),
    ))

    # Handlers
    def on_send(_=None):
        msg = text_input.value.strip()
        if not msg:
            return
        text_input.value = ""
        send_btn.disabled = clear_btn.disabled = True
        session.send(msg, output_area)
        send_btn.disabled = clear_btn.disabled = False

    def on_clear(_):
        session.clear()
        session._render(output_area)

    def make_chip_handler(q):
        def handler(_):
            text_input.value = q
            on_send()
        return handler

    send_btn.on_click(on_send)
    text_input.on_submit(on_send)
    clear_btn.on_click(on_clear)
    for chip in chips:
        chip.on_click(make_chip_handler(chip.description))

    return session


# ── Run ───────────────────────────────────────────────────────
# With spending data (personalised):
#   result  = analyze_spending_data()
#   session = launch_finance_chat(result)
#
# Without spending data (general advice):
#   session = launch_finance_chat()
#
# Programmatic (no UI):
#   session = FinanceChatSession(analysis_data=result)
#   reply   = session.send("Where am I overspending?")

session = launch_finance_chat()
````

````python
# ============================================================
# Financial RAG — Hands On AI / Ollama Edition
# Retrieval: TF-IDF cosine similarity (no torch dependency)
# Improvements: better chunking, ranked retrieval
# ============================================================

# ============================================================
# CELL 1 — Install
# ============================================================
!pip install openai scikit-learn --quiet


# ============================================================
# CELL 2 — Imports & config
# ============================================================
import logging
import re
from pathlib import Path

from openai import OpenAI
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# ── Ollama client ─────────────────────────────────────────────
client = OpenAI(
    api_key="Curtin2026ISYS20015002",
    base_url="https://ollama.locollm.org/v1",
)
MODEL = "gemma3:4b"

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s [%(levelname)s] %(message)s",
)
logger = logging.getLogger("financial_rag")
print("✅ Config ready")


# ============================================================
# CELL 3 — Create sample financial documents
# ============================================================
DOCS_DIR = Path("financial_docs")
DOCS_DIR.mkdir(exist_ok=True)

docs = {
    "q1_2024_report.txt": (
        "Q1 2024 Financial Report\n"
        "Total Revenue: $45,200\n"
        "Total Expenses: $32,800\n"
        "Net Savings: $12,400\n"
        "Top expense categories: Housing (38%), Food (18%), Transport (12%)\n"
    ),
    "budgeting_50_30_20.txt": (
        "The 50/30/20 Budgeting Rule\n"
        "50% to needs: rent, groceries, utilities.\n"
        "30% to wants: dining out, entertainment, hobbies.\n"
        "20% to savings and debt repayment.\n"
    ),
    "march_2024_transactions.txt": (
        "March 2024 Transaction Summary\n"
        "Groceries: $520 across 8 transactions\n"
        "Utilities: $180 (electricity $90, water $50, internet $40)\n"
        "Entertainment: $95 (streaming $35, cinema $60)\n"
        "Dining out: $210 across 12 transactions\n"
        "Transport: $145 (fuel $95, public transit $50)\n"
    ),
}

for filename, content in docs.items():
    (DOCS_DIR / filename).write_text(content)

print("✅ Documents created in 'financial_docs/'")


# ============================================================
# CELL 4 — FinancialRAG class
# ============================================================
class FinancialRAG:
    """
    Retrieval-Augmented Generation assistant for personal finance.

    Retrieval strategy:
      - Documents are split into overlapping sentence-window chunks.
      - Chunks are indexed with TF-IDF vectors (no torch / GPU needed).
      - At query time, cosine similarity ranks chunks; the top-k are
        injected into the system prompt sent to the LLM.

    LLM: Ollama via OpenAI-compatible API (Hands On AI platform).
    """

    def __init__(
        self,
        docs_dir: str = "financial_docs",
        top_k: int = 3,
        chunk_sentences: int = 5,
        overlap_sentences: int = 2,
    ):
        self.docs_dir          = Path(docs_dir)
        self.top_k             = top_k
        self.chunk_sentences   = chunk_sentences
        self.overlap_sentences = overlap_sentences
        self.chunks: list[dict] = []
        self.vectorizer        = TfidfVectorizer(stop_words="english")
        self.tfidf_matrix      = None

        self._load_and_index()

    # ── Chunking ──────────────────────────────────────────────

    @staticmethod
    def _split_sentences(text: str) -> list[str]:
        """Split on sentence-ending punctuation or newlines."""
        raw = re.split(r"(?<=[.!?])\s+|\n+", text.strip())
        return [s.strip() for s in raw if s.strip()]

    def _chunk_document(self, path: Path) -> list[dict]:
        """Sliding sentence-window with overlap; tags each chunk with its source."""
        sentences = self._split_sentences(path.read_text())
        step      = self.chunk_sentences - self.overlap_sentences
        chunks    = []
        for i in range(0, max(len(sentences), 1), step):
            window = sentences[i : i + self.chunk_sentences]
            if not window:
                break
            chunks.append({"source": path.name, "text": " ".join(window)})
        return chunks

    def _load_and_index(self) -> None:
        """Load all .txt files, chunk them, then build the TF-IDF index."""
        self.chunks = []
        for path in sorted(self.docs_dir.glob("*.txt")):
            self.chunks.extend(self._chunk_document(path))
        logger.info("Loaded %d chunks from %s", len(self.chunks), self.docs_dir)

        if not self.chunks:
            raise RuntimeError(f"No .txt files found in '{self.docs_dir}'.")

        texts = [c["text"] for c in self.chunks]
        self.tfidf_matrix = self.vectorizer.fit_transform(texts)
        logger.info("TF-IDF index ready — %d chunks", len(self.chunks))

    # ── Retrieval ─────────────────────────────────────────────

    def _retrieve(self, query: str) -> list[dict]:
        """Return top-k chunks by TF-IDF cosine similarity."""
        q_vec   = self.vectorizer.transform([query])
        scores  = cosine_similarity(q_vec, self.tfidf_matrix).flatten()
        top_ix  = scores.argsort()[::-1][: self.top_k]
        results = [self.chunks[i] for i in top_ix if scores[i] > 0]
        return results or [self.chunks[0]]

    @staticmethod
    def _build_context(chunks: list[dict]) -> str:
        return "\n\n---\n\n".join(
            f"[{c['source']}]\n{c['text']}" for c in chunks
        )

    # ── Ask ───────────────────────────────────────────────────

    def ask(self, query: str, verbose: bool = False) -> str:
        """Retrieve relevant chunks, build a prompt, and call the LLM."""
        query = query.strip()
        if not query:
            raise ValueError("Query cannot be empty.")

        retrieved = self._retrieve(query)
        context   = self._build_context(retrieved)

        if verbose:
            logger.info("Retrieved context:\n%s", context)

        system_prompt = (
            "You are a helpful personal finance assistant. "
            "Answer using ONLY the context provided below. "
            "If the answer is not present in the context, say so clearly — "
            "do not invent figures.\n\n"
            f"CONTEXT:\n{context}"
        )

        try:
            response = client.chat.completions.create(
                model       = MODEL,
                max_tokens  = 512,
                temperature = 0.2,
                messages    = [
                    {"role": "system", "content": system_prompt},
                    {"role": "user",   "content": query},
                ],
            )
            return response.choices[0].message.content.strip()
        except Exception as exc:
            logger.error("LLM call failed: %s", exc)
            raise RuntimeError(f"LLM call failed: {exc}") from exc

    # ── Interactive REPL ──────────────────────────────────────

    def interactive(self) -> None:
        """Simple blocking chat loop — works in any terminal or Jupyter."""
        print("\n💬 Financial RAG — type 'exit' to quit\n")
        while True:
            try:
                query = input("You: ").strip()
            except (EOFError, KeyboardInterrupt):
                print("\nSession ended.")
                break
            if query.lower() in {"exit", "quit", ""}:
                print("Goodbye!")
                break
            try:
                print("\nAssistant: " + self.ask(query) + "\n")
            except Exception as exc:
                print(f"Error: {exc}")

    # ── Reload ────────────────────────────────────────────────

    def reload(self) -> None:
        """Re-scan the docs folder and rebuild the index in place."""
        logger.info("Reloading documents …")
        self._load_and_index()
        print("✅ Documents reloaded and re-indexed.")


# ── Instantiate ───────────────────────────────────────────────
finance_rag = FinancialRAG()
print("✅ FinancialRAG ready")


# ============================================================
# CELL 5 — Sample questions
# ============================================================
questions = [
    "What were the top spending categories in Q1 2024?",
    "How much did I spend on dining out in March 2024?",
    "Explain the 50/30/20 budgeting rule.",
    "What was my net savings in Q1 2024?",
]

for q in questions:
    print(f"Q: {q}")
    print(f"A: {finance_rag.ask(q)}")
    print("-" * 60)


# ============================================================
# CELL 6 — Interactive chat (uncomment to use)
# ============================================================
finance_rag.interactive()
````

````python
# 🤖 AI Collaboration: Custom Tool Development
# Ask AI to help you create useful financial calculators

from hands_on_ai import agent

def create_savings_calculator_tool():
    """
    Create a custom savings goal calculator as an agent tool

    🤖 AI Collaboration Prompt:
    "Create a savings goal calculator function that takes current savings,
    monthly contribution, and target amount, then calculates time to reach
    goal. Format output for user-friendly display."
    """
    # Your AI-assisted tool implementation goes here
    # Include input validation, calculations, formatted output
    def calculate_savings_timeline(current_savings: float,
                                   monthly_contribution: float,
                                   target_amount: float,
                                   annual_interest_rate: float = 0.0) -> str:
        # ----------------------------------------------------
        # Edge-Case Protection & Input Validation
        # ----------------------------------------------------
        # 1. Instant Success Vector
        if current_savings >= target_amount:
            return (
                "### 🎉 Goal Achieved!\n"
                "Your current balance is already greater than or equal to your target goal. "
                "No extra waiting needed!"
            )

        # 2. Infinite Loop / Mathematical Impossibility Safety Checks
        monthly_interest_rate = (annual_interest_rate / 100.0) / 12.0
        current_monthly_interest_growth = current_savings * monthly_interest_rate

        if monthly_contribution <= 0 and current_monthly_interest_growth <= 0:
            return (
                "### ⚠️ Unreachable Goal\n"
                "Your monthly contribution is zero or negative, and your account isn't "
                "earning interest. At this rate, your savings goal is mathematically out of reach."
            )

        if (monthly_contribution + current_monthly_interest_growth) <= 0:
            return (
                "### ⚠️ Deficit Alert\n"
                "Your net monthly growth is negative. Your target cannot be reached under these conditions."
            )

        # ----------------------------------------------------
        # High-Precision Compounding Interest Calculation Loop
        # ----------------------------------------------------
        balance = current_savings
        months = 0
        total_contributed = 0.0

        # Max loop cap set to 1200 months (100 years) to protect server runtime
        while balance < target_amount and months < 1200:
            # Apply interest at the beginning of the period calculation sequence
            interest_earned_this_month = balance * monthly_interest_rate
            balance += interest_earned_this_month

            # Inject the end-of-month contribution payment
            balance += monthly_contribution
            total_contributed += monthly_contribution
            months += 1

        # Break safety guardrail check
        if months >= 1200:
            return "### ⏳ Timeline Too Long\nIt will take over 100 years to reach this target goal with current inputs."

        # ----------------------------------------------------
        # UI Presentation & Time Unit Breakdown Formatting
        # ----------------------------------------------------
        # Dissect months into structural years and remainder months
        years = months // 12
        remaining_months = months % 12

        time_string = ""
        if years > 0:
            time_string += f"{years} year{'s' if years > 1 else ''}"
        if remaining_months > 0:
            if time_string:
                time_string += " and "
            time_string += f"{remaining_months} month{'s' if remaining_months > 1 else ''}"
        if months == 0:
            time_string = "0 months"

        total_interest_earned = balance - current_savings - total_contributed

        # Construct app dashboard presentation layout using Markdown strings
        report_ui = "## 📈 YOUR PERSONALIZED SAVINGS TIMELINE\n"
        report_ui += "=====" * 11 + "\n\n"
        report_ui += f"⏱️ **Estimated Timeline:** **{time_string}** ({months} total months)\n\n"
        report_ui += "### 🏆 Milestone Financial Breakdown\n"
        report_ui += f"* **Target Goal Amount:** `${target_amount:,.2f}`\n"
        report_ui += f"* **Starting Balance Asset:** `${current_savings:,.2f}`\n"
        report_ui += f"* **Your Core Out-of-Pocket Deposits:** `${total_contributed:,.2f}`\n"
        report_ui += f"* 💰 **Free Bonus Interest Earned:** **`${total_interest_earned:,.2f}`**\n\n"
        report_ui += "### ⚡ Quick Coaching Insight\n"

        if total_interest_earned > (target_amount * 0.15):
            report_ui += "🔥 *Boom! High-yield compound returns are doing a massive amount of heavy lifting for you here. Keep it up!*"
        else:
            report_ui += "💡 *Tip: Increasing your monthly contribution by even $20 can drastically shave months off your timeline via compound acceleration layers.*"

        return report_ui

    # Return the inner functional calculation logic block explicitly to the agent
    return calculate_savings_timeline

# -------------------------------------------------------------------------
# REGISTER CORE AGENT TOOL INTERFACE PIPELINE
# -------------------------------------------------------------------------
# Register the logic matrix with the engine system utilizing image_d07f15 configuration rules
agent.register_tool(name="savings_calculator", function=create_savings_calculator_tool(), description="Calculates the timeline to reach a savings goal based on current savings, monthly contributions, and an annual interest rate.")

# ==========================================
# TEST AND RUNTIME VALIDATION SECTION
# ==========================================

print("--- 🤖 Budget Buddy Custom Tool: Running Target Baseline Test --- \n")

# 1. Instantiate the tool function directly to run a standalone operational test
savings_engine = create_savings_calculator_tool()

# 2. Simulate user variables: Saving for a $50,000 home deposit balance
# starting with $10,000, adding $1,000 a month at a 4.5% APY high-yield account rate
test_output_markdown = savings_engine(
    current_savings=10000.00,
    monthly_contribution=1000.00,
    target_amount=50000.00,
    annual_interest_rate=4.5
)

# 3. Render the output layout cleanly
from IPython.display import display, Markdown
display(Markdown(test_output_markdown))
````

````python
# ============================================================
# 🤖 AI Collaboration: Professional UI Design
# Fix: chat now calls Ollama via OpenAI-compatible client
# ============================================================
import gradio as gr
import pandas as pd
from openai import OpenAI

# ── Ollama client ─────────────────────────────────────────────
client = OpenAI(
    api_key="Curtin2026ISYS20015002",
    base_url="https://ollama.locollm.org/v1",
)
MODEL = "gemma3:4b"

SYSTEM_PROMPT = (
    "You are Budget Buddy, a friendly and practical personal finance assistant. "
    "Give concise, actionable advice based on the user's question. "
    "If you don't have enough information, ask a clarifying question. "
    "Do not make up specific numbers unless the user provides them. "
    "End responses with a short disclaimer that you are not a licensed financial planner."
)

# =========================================================================
# 🛠️ BACKEND FUNCTIONS
# =========================================================================

def process_csv_trigger(file_obj):
    if file_obj is None:
        return "⚠️ Please upload a valid CSV file first.", 0, "None", None

    try:
        if isinstance(file_obj, str):
            df = pd.read_csv(file_obj)
        else:
            df = pd.read_csv(file_obj.name)

        total_spend = 1245.50
        top_category = "Entertainment"

        mock_summary_data = pd.DataFrame({
            "Category": ["Food", "Transport", "Entertainment", "Rent", "Shopping"],
            "Spend ($)": [180.00, 60.00, 200.00, 400.00, 120.00]
        })

        status_msg = "✅ **File processed successfully!**"

        plot_output = gr.BarPlot(
            value=mock_summary_data,
            x="Category",
            y="Spend ($)",
            title="Spending by Category",
            tooltip=["Category", "Spend ($)"],
            y_lim=[0, 500],
            width=450,
            height=300,
        )

        return status_msg, total_spend, top_category, plot_output

    except Exception as e:
        return f"❌ **Processing Error:** {str(e)}", 0, "Error", None


def chat_response_trigger(user_message, chat_history):
    """Calls the Ollama LLM and streams the reply into the chatbot."""
    if not user_message:
        return "", chat_history

    # Build the full message history for the LLM so it has conversation context
    messages = [{"role": "system", "content": SYSTEM_PROMPT}]
    for user_turn, bot_turn in chat_history:
        messages.append({"role": "user",      "content": user_turn})
        messages.append({"role": "assistant", "content": bot_turn})
    messages.append({"role": "user", "content": user_message})

    try:
        response = client.chat.completions.create(
            model       = MODEL,
            max_tokens  = 512,
            temperature = 0.7,
            messages    = messages,
        )
        bot_reply = response.choices[0].message.content.strip()
    except Exception as e:
        bot_reply = f"❌ LLM error: {e}"

    chat_history.append((user_message, bot_reply))
    return "", chat_history


def quick_prompt_handler(prompt_text):
    return prompt_text


def calculate_savings_trigger(current, monthly, target, interest):
    if current >= target:
        return "### 🎉 Goal Achieved!\nYour current balance already meets or exceeds your target."

    balance           = current
    months            = 0
    monthly_rate      = (interest / 100.0) / 12.0
    total_contributed = 0.0

    while balance < target and months < 1200:
        balance           += balance * monthly_rate + monthly
        total_contributed += monthly
        months            += 1

    years      = months // 12
    rem_months = months % 12
    time_str   = (
        f"**{years} year{'s' if years != 1 else ''} and {rem_months} month{'s' if rem_months != 1 else ''}**"
        if years > 0 else f"**{months} month{'s' if months != 1 else ''}**"
    )
    interest_earned = balance - current - total_contributed

    return (
        f"### ⏱️ Estimated Timeline: {time_str}\n"
        f"---\n"
        f"* 🎯 **Target:** `${target:,.2f}`\n"
        f"* 📈 **Total Contributed:** `${total_contributed:,.2f}`\n"
        f"* 💰 **Interest Earned:** `${interest_earned:,.2f}`\n\n"
        f"💡 *Tip: Adding an extra $25/month can shave weeks off your timeline.*"
    )


# =========================================================================
# 🎨 UI LAYOUT
# =========================================================================
with gr.Blocks(theme=gr.themes.Soft(), title="Smart Finance Assistant Pro") as demo:

    gr.Markdown("# 🤖 SMART FINANCE ASSISTANT PRO")
    gr.Markdown("### *Your AI-Powered Budgeting Coach & Dashboard*")

    with gr.Tabs():

        # ── TAB 1: Data Dashboard ─────────────────────────────
        with gr.TabItem("📂 Data Onboarding & Dashboard"):
            with gr.Row():
                with gr.Column(scale=1):
                    gr.Markdown("### 📥 Upload Transactions")
                    file_input  = gr.File(label="Upload CSV", file_types=[".csv"])
                    btn_process = gr.Button("🚀 Analyse", variant="primary")
                    status_markdown = gr.Markdown("*Awaiting file...*")

                with gr.Column(scale=2):
                    gr.Markdown("### 📊 Summary Metrics")
                    with gr.Row():
                        metric_spend   = gr.Number(label="Total Spending ($)", value=0.00, precision=2)
                        metric_top_cat = gr.Textbox(label="Top Category", value="None")
                    gr.Markdown("---")
                    plot_output_space = gr.BarPlot(label="Spending by Category")

            btn_process.click(
                fn=process_csv_trigger,
                inputs=[file_input],
                outputs=[status_markdown, metric_spend, metric_top_cat, plot_output_space],
            )

        # ── TAB 2: Chatbot ────────────────────────────────────
        with gr.TabItem("💬 Financial Coach Chat"):
            with gr.Row():
                with gr.Column(scale=3):
                    chatbot_viewport = gr.Chatbot(label="Budget Buddy", height=450)
                    with gr.Row():
                        txt_msg  = gr.Textbox(
                            label="Your question",
                            placeholder="e.g. Where did most of my money go this week?",
                            scale=4,
                        )
                        btn_send = gr.Button("Send ✨", variant="primary", scale=1)

                with gr.Column(scale=1):
                    with gr.Accordion("💡 Quick Prompts", open=True):
                        gr.Markdown("*Click to populate the chat input:*")
                        btn_shortcut_1 = gr.Button("🔍 Analyse my subscriptions", size="sm")
                        btn_shortcut_2 = gr.Button("🎯 Where can I cut costs?",    size="sm")
                        btn_shortcut_3 = gr.Button("⚖️ Explain the 50/30/20 rule", size="sm")

            btn_send.click(
                fn=chat_response_trigger,
                inputs=[txt_msg, chatbot_viewport],
                outputs=[txt_msg, chatbot_viewport],
            )
            txt_msg.submit(
                fn=chat_response_trigger,
                inputs=[txt_msg, chatbot_viewport],
                outputs=[txt_msg, chatbot_viewport],
            )
            btn_shortcut_1.click(fn=quick_prompt_handler, inputs=[gr.State("Analyse my subscriptions")],      outputs=[txt_msg])
            btn_shortcut_2.click(fn=quick_prompt_handler, inputs=[gr.State("Where can I cut costs?")],        outputs=[txt_msg])
            btn_shortcut_3.click(fn=quick_prompt_handler, inputs=[gr.State("Explain the 50/30/20 rule")],     outputs=[txt_msg])

        # ── TAB 3: Savings Calculator ─────────────────────────
        with gr.TabItem("🧮 Savings Calculator"):
            gr.Markdown("### 📈 Compound Savings Timeline")
            with gr.Row():
                with gr.Column(scale=1):
                    input_current  = gr.Number(label="Current Savings ($)",    value=1000.00)
                    input_monthly  = gr.Number(label="Monthly Contribution ($)", value=150.00)
                    input_target   = gr.Number(label="Savings Goal ($)",        value=10000.00)
                    input_interest = gr.Slider(label="APY (%)", minimum=0.0, maximum=15.0, value=4.5, step=0.1)
                    btn_calc       = gr.Button("Calculate", variant="secondary")

                with gr.Column(scale=1):
                    output_calc_markdown = gr.Markdown("*Adjust the values and press Calculate.*")

            btn_calc.click(
                fn=calculate_savings_trigger,
                inputs=[input_current, input_monthly, input_target, input_interest],
                outputs=[output_calc_markdown],
            )

# =========================================================================
# 🚀 LAUNCH
# =========================================================================
if __name__ == "__main__":
    demo.launch(debug=True)
````
