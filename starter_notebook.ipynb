### Foundation Data Processing Functions

````python
import pandas as pd
import io

# 🤖 AI Collaboration: Data Loading and Cleaning Function
# Ask AI to help you create a robust data loading function

def load_and_clean_transaction_data(file_path):
    """
    Load and clean transaction data for the Smart Finance Assistant

    🤖 AI Collaboration Prompt:
    "Create a function to load CSV transaction data with Date, Amount, Category,
    Description columns. Handle dollar signs in Amount, missing values, and
    data validation. Include clear business-focused error messages."

    Args:
        file_path: Path to CSV file or file object
    Returns:
        pandas.DataFrame: Cleaned transaction data
    """
    required_columns = {'Date', 'Amount', 'Category', 'Description'}

    try:
        # Step 1: Load the CSV file
        # supports both string paths and file-like objects (like io.StringIO)
        df = pd.read_csv(file_path)

        if df.empty:
            raise ValueError("The uploaded file contains no data rows.")

        # Step 2: Validate that all required column headers exist
        missing_cols = required_columns - set(df.columns)
        if missing_cols:
            raise KeyError(f"Missing required columns: {', '.join(missing_cols)}. Please ensure Date, Amount, Category, and Description are present.")

        # Step 3: Handle Missing Values
        # Drop rows where crucial indexing data (Date or Amount) is null
        df = df.dropna(subset=['Date', 'Amount'])
        # Fill non-critical structural fields with 'Unknown'
        df['Category'] = df['Category'].fillna('Unknown').astype(str).str.strip()
        df['Description'] = df['Description'].fillna('Unknown').astype(str).str.strip()

        # Step 4: Clean and Parse 'Amount' Column
        # Cast to string first to use string vector methods safely
        df['Amount'] = df['Amount'].astype(str).str.replace('$', '', regex=False)
        df['Amount'] = df['Amount'].str.replace(',', '', regex=False)
        df['Amount'] = df['Amount'].str.replace('(', '-', regex=False).str.replace(')', '', regex=False)
        df['Amount'] = df['Amount'].str.strip()

        # Convert to float and force invalid text formats to NaN
        df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')

        # Step 5: Clean and Parse 'Date' Column
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

        # Drop any records that failed numeric or datetime coercion conversions
        df = df.dropna(subset=['Date', 'Amount'])

        # Re-verify DataFrame isn't empty after rigorous data scrubbing
        if df.empty:
            raise ValueError("Data validation failed: No valid records remained after cleaning numeric and date formats.")

        # Reset the index cleanly for subsequent RAG operations
        return df.reset_index(drop=True)

    except FileNotFoundError:
        raise Exception("Error: The financial transaction file could not be located at the provided path.")
    except pd.errors.EmptyDataError:
        raise Exception("Error: The uploaded CSV file is blank or corrupted.")
    except (KeyError, ValueError) as business_err:
        raise Exception(f"Business Validation Error: {str(business_err)}")
    except Exception as general_err:
        raise Exception(f"An unexpected system exception occurred during file processing: {str(general_err)}")

# ==========================================
# TEST AND VALIDATION SECTION
# ==========================================

# 1. Mocking a successful CSV string layout matching your Step 3 data
valid_csv_data = """Date,Amount,Category,Description
2026-05-04,$45,Food,Lunch
2026-05-05,$30,Transport,Bus Pass
2026-05-06,$75,Food,Dinner
2026-05-07,$100,Shopping,Clothes
2026-05-08,-$40,Refund,Returned item
2026-05-09,($15.50),Food,Late Night Fast Food
, $20, Food, Missing Date Row (Should drop)
2026-05-10,,Shopping,Missing Amount Row (Should drop)
2026-05-11,$12,,Missing Category Row (Should become Unknown)
"""

print("--- Test Case 1: Processing Valid Data Ingest ---")
try:
    # io.StringIO tricks pandas into treating a string exactly like a physical file
    clean_df = load_and_clean_transaction_data(io.StringIO(valid_csv_data))
    print("\nSuccessfully loaded and transformed DataFrame:")
    print(clean_df.to_string())
    print(f"\nDate column dtype structure: {clean_df['Date'].dtype}")
    print(f"Amount column dtype structure: {clean_df['Amount'].dtype}")
except Exception as e:
    print(e)

print("\n" + "="*50 + "\n")

# 2. Mocking a failed layout with incorrect metadata columns
invalid_csv_data = """TransactionDate,CashValue,Tag
2026-05-04,$45,Food
2026-05-05,$30,Transport
"""

print("--- Test Case 2: Triggering Business Validation Error ---")
try:
    bad_df = load_and_clean_transaction_data(io.StringIO(invalid_csv_data))
except Exception as e:
    print(e)
````

````python
# 🤖 AI Collaboration: Spending Analysis Function
# Ask AI to help you create comprehensive spending analysis

def analyze_spending_patterns(df):
    """
    Analyze spending patterns and generate business insights

    🤖 AI Collaboration Prompt:
    "Create a function that analyzes spending by category, calculates percentages,
    identifies top spending areas, and generates actionable financial insights
    formatted for business presentation."

    Args:
        df: Cleaned transaction DataFrame
    Returns:
        dict: Analysis results and insights
    """
    # Your AI-assisted implementation goes here
    # ----------------------------------------------------
    # Business Logic: Separate Positive Outflows from Refunds
    # ----------------------------------------------------
    # Filter out positive outlays to evaluate actual cost structures
    spending_df = df[df['Amount'] > 0].copy()
    refund_df = df[df['Amount'] < 0].copy()

    total_gross_spend = spending_df['Amount'].sum()
    total_refunds = refund_df['Amount'].sum()
    net_spend = df['Amount'].sum()

    if total_gross_spend == 0:
        return {
            "summary_text": "### 📊 Financial Report Summary\nNo positive spending records detected in this period.",
            "metrics": {}
        }

    # ----------------------------------------------------
    # Step 1 & 2: Aggregation and Proportional Analysis
    # ----------------------------------------------------
    category_group = spending_df.groupby('Category')['Amount'].agg(['sum', 'count'])
    category_group = category_group.rename(columns={'sum': 'Total Spend', 'count': 'Transaction Count'})

    # Calculate the percentage each category contributes to total gross spending
    category_group['Percentage'] = (category_group['Total Spend'] / total_gross_spend) * 100
    category_group = category_group.sort_values(by='Total Spend', ascending=False)

    # Isolate top spending areas
    top_categories = category_group.index.tolist()[:3]

    # ----------------------------------------------------
    # Step 3: Actionable Financial Insights Engine
    # ----------------------------------------------------
    insights = []

    # Business Insight rule 1: Fixed/Recurring cost threshold check
    fixed_categories = ['Utilities', 'Rent', 'Insurance', 'Bills']
    fixed_spend = category_group[category_group.index.isin(fixed_categories)]['Total Spend'].sum()
    fixed_percentage = (fixed_spend / total_gross_spend) * 100

    if fixed_percentage > 50:
        insights.append(f"⚠️ **Fixed Cost Warning:** Critical structural commitments ({fixed_percentage:.1f}% of total spend) exceed the 50% healthy financial framework baseline. Review underlying subscription or utility plans for structural cost reduction.")
    else:
        insights.append(f"✅ **Fixed Cost Allocation:** Structural commitments are healthy at {fixed_percentage:.1f}% of overall outlays, leaving room for savings and flexible adjustments.")

    # Business Insight rule 2: Discretionary/Spike Category tracking
    discretionary_categories = ['Food', 'Dining Out', 'Shopping', 'Entertainment', 'Retail']
    for cat in discretionary_categories:
        if cat in category_group.index:
            cat_pct = category_group.loc[cat, 'Percentage']
            # Highlight if a single discretionary bucket devours more than 25% of all funds
            if cat_pct > 25:
                insights.append(f"🚨 **Discretionary Spending Spike:** **{cat}** represents a disproportionate **{cat_pct:.1f}%** of your entire operational spending layout. Consider implementing a temporary soft freeze on this category.")

    # Business Insight rule 3: Refund Optimization Impact
    if abs(total_refunds) > 0:
        refund_efficiency = (abs(total_refunds) / total_gross_spend) * 100
        insights.append(f"ℹ️ **Refund Adjustments:** Validated returns recovered **${abs(total_refunds):,.2f}** back to operational capital, mitigating gross spend metrics by **{refund_efficiency:.1f}%**.")

    # ----------------------------------------------------
    # Step 4: Presentation Presentation Formatting Layout
    # ----------------------------------------------------
    report_str = "## 📊 SMART FINANCE ASSISTANT: STRATEGIC BUSINESS REPORT\n"
    report_str += "=====" * 11 + "\n\n"

    report_str += "### 💎 Core Aggregated Metrics\n"
    report_str += f"* **Gross Capital Outflow:** `${total_gross_spend:,.2f}`\n"
    report_str += f"* **Total Recovered Capital (Refunds):** `${abs(total_refunds):,.2f}`\n"
    report_str += f"* **Net Financial Position:** `${net_spend:,.2f}`\n\n"

    report_str += "### 📂 Categorized Expenditure Matrix\n"
    report_str += "| Category | Total Spend | Vol (Count) | % of Total Budget |\n"
    report_str += "| :--- | :--- | :--- | :--- |\n"

    for cat, row in category_group.iterrows():
        report_str += f"| **{cat}** | ${row['Total Spend']:,.2f} | {int(row['Transaction Count'])}x | {row['Percentage']:.1f}% |\n"

    report_str += f"\n🔥 **Primary Exposure Vectors (Top 3 Heavy Sectors):** {', '.join(top_categories)}\n\n"

    report_str += "### 💡 Strategic Actionable Insights\n"
    for insight in insights:
        report_str += f"* {insight}\n"

    # Packaging both clean structural dict components and visual text for UI display compatibility
    return {
        "summary_markdown": report_str,
        "raw_matrix": category_group,
        "top_sectors": top_categories
    }

# ==========================================
# TEST AND RUNTIME VALIDATION SECTION
# ==========================================

# 1. Let's pull the clean data output variables generated in your previous step
try:
    # Running analysis on the clean_df generated in our previous functional block
    analysis_results = analyze_spending_patterns(clean_df)

    # Display the business-formatted presentation string output directly
    from IPython.display import display, Markdown
    display(Markdown(analysis_results["summary_markdown"]))

except NameError:
    print("Previous dataframe variables weren't located in this runtime environment. Generating standalone verification matrix alternative:")
    # Fallback simulation database block matching your manual Step 3 calculations
    fallback_data = pd.DataFrame([
        {"Date": "2026-05-01", "Amount": 120.00, "Category": "Food", "Description": "Groceries"},
        {"Date": "2026-05-02", "Amount": 60.00, "Category": "Transport", "Description": "Taxi"},
        {"Date": "2026-05-03", "Amount": 200.00, "Category": "Entertainment", "Description": "Concert Ticket"},
        {"Date": "2026-05-04", "Amount": 400.00, "Category": "Rent", "Description": "Studio Unit Bills"},
        {"Date": "2026-05-05", "Amount": -50.00, "Category": "Refund", "Description": "Returned item Outlay"}
    ])
    # Force convert types to simulate authentic load pipelines
    fallback_data['Date'] = pd.to_datetime(fallback_data['Date'])

    analysis_results = analyze_spending_patterns(fallback_data)
    from IPython.display import display, Markdown
    display(Markdown(analysis_results["summary_markdown"]))
````

````python
# 🤖 AI Collaboration: Business Insights Generator
# Ask AI to help create professional financial recommendations

def generate_financial_recommendations(analysis_data):
    """
    Generate actionable financial recommendations based on spending analysis

    🤖 AI Collaboration Prompt:
    "Based on spending analysis data, create professional financial
    recommendations. Include specific savings opportunities, spending
    pattern observations, and actionable advice formatted for a
    personal finance app user."

    Args:
        analysis_data: Dictionary with spending analysis results
    Returns:
        str: Formatted recommendations report
    """
    # Your AI-assisted implementation goes here
    # ----------------------------------------------------
    # Data Extraction & Fallback Protection
    # ----------------------------------------------------
    # Extract data whether it's passed as a raw dictionary or our previous analysis object
    if isinstance(analysis_data, dict) and "raw_matrix" in analysis_data:
        matrix = analysis_data["raw_matrix"]
        category_totals = matrix['Total Spend'].to_dict()
    elif isinstance(analysis_data, dict):
        category_totals = analysis_data
    else:
        return "### ❌ System Alert\nUnable to read your transaction matrix formatting profile."

    # Calculate total positive spending for baseline proportions
    gross_spend = sum(val for val in category_totals.values() if val > 0)

    if gross_spend == 0:
        return "## 🌟 Welcome to Budget Buddy!\nStart tracking your expenses to unlock personalized behavioral insights and quick wins."

    # Safe data grabber helper function
    def get_pct(category_name):
        amount = category_totals.get(category_name, 0)
        return (amount / gross_spend) * 100 if amount > 0 else 0

    # ----------------------------------------------------
    # Section 1: Spending Pattern Observations (The "What")
    # ----------------------------------------------------
    observations = []

    # Check Food / Dining behaviors
    food_pct = get_pct('Food') + get_pct('Dining Out')
    if food_pct > 30:
        observations.append(f"🍔 **Fueling the food scene:** Delicious food and quick treats made up **{food_pct:.1f}%** of your total spending. Your tastebuds are living large!")
    else:
        observations.append("🛒 **Kitchen champion:** You're doing an amazing job keeping your meal costs balanced and under control this period!")

    # Check Entertainment / Shopping leaks
    fun_pct = get_pct('Entertainment') + get_pct('Shopping') + get_pct('Retail')
    if fun_pct > 20:
        observations.append(f"🎉 **Living your best life:** Discretionary shopping and social entertainment claimed **{fun_pct:.1f}%** of your funds. It’s great to enjoy yourself, but let's see if we can optimize those leaks!")

    # Standard fallback if no heavy spikes occur
    if not observations:
        observations.append("📊 **Steady as she goes:** Your money is flowing evenly across your usual categories without any unusual spikes.")

    # ----------------------------------------------------
    # Section 2: Specific Savings Opportunities (The "Where")
    # ----------------------------------------------------
    savings_opportunities = []

    # 50/30/20 Rule alignment: Check Variable Wants vs Fixed Needs
    wants_spend = category_totals.get('Entertainment', 0) + category_totals.get('Shopping', 0) + category_totals.get('Dining Out', 0)
    wants_pct = (wants_spend / gross_spend) * 100 if wants_spend > 0 else 0

    if wants_pct > 30:
        savings_opportunities.append(f"⚖️ **The 50/30/20 Balance:** Your flexible 'Wants' are sitting at **{wants_pct:.1f}%** (the recommended target is under 30%). Trimming minor delivery or retail purchases is your fastest ticket to extra savings.")

    # Target specific grocery/subscription wins
    if 'Food' in category_totals and category_totals['Food'] > 100:
        savings_opportunities.append("🌾 **The Supermarket Swap:** Shifting from premium brands to local alternatives at Woolworths or Coles can instantly shave roughly 15% off your next bill.")

    if 'Entertainment' in category_totals:
        savings_opportunities.append("🍿 **Subscription Audit:** Take 2 minutes to check for overlapping active streaming platforms (Netflix, Stan, Spotify). Canceling just one unused tier keeps money directly in your pocket.")

    # ----------------------------------------------------
    # Section 3: Actionable Advice (The "How")
    # ----------------------------------------------------
    # Curate high-impact micro-actions for mobile screens
    actionable_wins = [
        "🔒 **Try a 48-Hour 'No-Spend' Challenge:** Freeze all discretionary shopping for the next two days. It’s an easy, gamified way to keep an extra $40–$60 safe in your account.",
        "📱 **Move Money First:** Next time you get paid, immediately auto-transfer 10% into a separate high-yield savings vault before you have a chance to spot it in your main spending balance.",
        "☕ **The 'Pause Before Purchase' Rule:** Leave item options sitting in your online shopping cart for 24 hours before checking out. It cuts down impulsive purchases by over 70%!"
    ]

    # ----------------------------------------------------
    # UI Presentation Formatting
    # ----------------------------------------------------
    ui_markdown = "## 💡 BUDGET BUDDY: YOUR PERSONALIZED COACHING DASHBOARD\n"
    ui_markdown += "Hi there! Let's take a quick look at your money habits and unlock some easy wins:\n\n"

    ui_markdown += "### 🔍 Did you notice?\n"
    for obs in observations:
        ui_markdown += f"* {obs}\n"
    ui_markdown += "\n"

    ui_markdown += "### 💎 Smart Savings Opportunities\n"
    for opp in savings_opportunities:
        ui_markdown += f"* {opp}\n"
    ui_markdown += "\n"

    ui_markdown += "### ⚡ Quick Micro-Wins (Try these today!)\n"
    for win in actionable_wins:
        ui_markdown += f"* {win}\n"

    return ui_markdown

# ==========================================
# TEST AND RUNTIME VALIDATION SECTION
# ==========================================

# Mock data mapping standard local consumer expenses
mock_australian_spend = {
    'Food': 180.00,          # Woolworths grocery run
    'Dining Out': 85.50,     # Grill'd burger nights + coffees
    'Entertainment': 65.00,  # Netflix tier + Concert Ticket layby
    'Transport': 45.00,      # Opal Card top-ups
    'Shopping': 120.00       # Retail clothing store purchase
}

print("--- Testing App Engine Notification Generation ---")

# Execute our behavioral UX function
coaching_payload = generate_financial_recommendations(mock_australian_spend)

# Display the mobile-friendly scannable Markdown output
from IPython.display import display, Markdown
display(Markdown(coaching_payload))
````

## Custom Financial Tools

````python
# 🤖 AI Collaboration: Custom Tool Development
# Ask AI to help you create useful financial calculators

from hands_on_ai import agent

def create_savings_calculator_tool():
    """
    Create a custom savings goal calculator as an agent tool

    🤖 AI Collaboration Prompt:
    "Create a savings goal calculator function that takes current savings,
    monthly contribution, and target amount, then calculates time to reach
    goal. Format output for user-friendly display."
    """
    # Your AI-assisted tool implementation goes here
    # Include input validation, calculations, formatted output
    def calculate_savings_timeline(current_savings: float,
                                   monthly_contribution: float,
                                   target_amount: float,
                                   annual_interest_rate: float = 0.0) -> str:
        # ----------------------------------------------------
        # Edge-Case Protection & Input Validation
        # ----------------------------------------------------
        # 1. Instant Success Vector
        if current_savings >= target_amount:
            return (
                "### 🎉 Goal Achieved!\n"
                "Your current balance is already greater than or equal to your target goal. "
                "No extra waiting needed!"
            )

        # 2. Infinite Loop / Mathematical Impossibility Safety Checks
        monthly_interest_rate = (annual_interest_rate / 100.0) / 12.0
        current_monthly_interest_growth = current_savings * monthly_interest_rate

        if monthly_contribution <= 0 and current_monthly_interest_growth <= 0:
            return (
                "### ⚠️ Unreachable Goal\n"
                "Your monthly contribution is zero or negative, and your account isn't "
                "earning interest. At this rate, your savings goal is mathematically out of reach."
            )

        if (monthly_contribution + current_monthly_interest_growth) <= 0:
            return (
                "### ⚠️ Deficit Alert\n"
                "Your net monthly growth is negative. Your target cannot be reached under these conditions."
            )

        # ----------------------------------------------------
        # High-Precision Compounding Interest Calculation Loop
        # ----------------------------------------------------
        balance = current_savings
        months = 0
        total_contributed = 0.0

        # Max loop cap set to 1200 months (100 years) to protect server runtime
        while balance < target_amount and months < 1200:
            # Apply interest at the beginning of the period calculation sequence
            interest_earned_this_month = balance * monthly_interest_rate
            balance += interest_earned_this_month

            # Inject the end-of-month contribution payment
            balance += monthly_contribution
            total_contributed += monthly_contribution
            months += 1

        # Break safety guardrail check
        if months >= 1200:
            return "### ⏳ Timeline Too Long\nIt will take over 100 years to reach this target goal with current inputs."

        # ----------------------------------------------------
        # UI Presentation & Time Unit Breakdown Formatting
        # ----------------------------------------------------
        # Dissect months into structural years and remainder months
        years = months // 12
        remaining_months = months % 12

        time_string = ""
        if years > 0:
            time_string += f"{years} year{'s' if years > 1 else ''}"
        if remaining_months > 0:
            if time_string:
                time_string += " and "
            time_string += f"{remaining_months} month{'s' if remaining_months > 1 else ''}"
        if months == 0:
            time_string = "0 months"

        total_interest_earned = balance - current_savings - total_contributed

        # Construct app dashboard presentation layout using Markdown strings
        report_ui = "## 📈 YOUR PERSONALIZED SAVINGS TIMELINE\n"
        report_ui += "=====" * 11 + "\n\n"
        report_ui += f"⏱️ **Estimated Timeline:** **{time_string}** ({months} total months)\n\n"
        report_ui += "### 🏆 Milestone Financial Breakdown\n"
        report_ui += f"* **Target Goal Amount:** `${target_amount:,.2f}`\n"
        report_ui += f"* **Starting Balance Asset:** `${current_savings:,.2f}`\n"
        report_ui += f"* **Your Core Out-of-Pocket Deposits:** `${total_contributed:,.2f}`\n"
        report_ui += f"* 💰 **Free Bonus Interest Earned:** **`${total_interest_earned:,.2f}`**\n\n"
        report_ui += "### ⚡ Quick Coaching Insight\n"

        if total_interest_earned > (target_amount * 0.15):
            report_ui += "🔥 *Boom! High-yield compound returns are doing a massive amount of heavy lifting for you here. Keep it up!*"
        else:
            report_ui += "💡 *Tip: Increasing your monthly contribution by even $20 can drastically shave months off your timeline via compound acceleration layers.*"

        return report_ui

    # Return the inner functional calculation logic block explicitly to the agent
    return calculate_savings_timeline

# -------------------------------------------------------------------------
# REGISTER CORE AGENT TOOL INTERFACE PIPELINE
# -------------------------------------------------------------------------
# Register the logic matrix with the engine system utilizing image_d07f15 configuration rules
agent.register_tool(name="savings_calculator", function=create_savings_calculator_tool(), description="Calculates the timeline to reach a savings goal based on current savings, monthly contributions, and an annual interest rate.")

# ==========================================
# TEST AND RUNTIME VALIDATION SECTION
# ==========================================

print("--- 🤖 Budget Buddy Custom Tool: Running Target Baseline Test --- \n")

# 1. Instantiate the tool function directly to run a standalone operational test
savings_engine = create_savings_calculator_tool()

# 2. Simulate user variables: Saving for a $50,000 home deposit balance
# starting with $10,000, adding $1,000 a month at a 4.5% APY high-yield account rate
test_output_markdown = savings_engine(
    current_savings=10000.00,
    monthly_contribution=1000.00,
    target_amount=50000.00,
    annual_interest_rate=4.5
)

# 3. Render the output layout cleanly
from IPython.display import display, Markdown
display(Markdown(test_output_markdown))
````
